## 실습문제 - 실리콘 밸리 2

이번에는 어도비 (Adobe)의 직원 분포를 한번 살펴봅시다.

어도비 전체 직원들의 직군 분포를 파이 그래프로 그려보세요.

(인원이 0인 직군은 그래프에 표시되지 않아야 합니다.)

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/silicon_valley_details.csv')

adobe = (df['company'] == 'Adobe') & (df['count'] != 0) & \
(df['job_category'] != 'Totals') & (df['job_category'] != 'Previous_totals') & \
(df['race'] == 'Overall_totals')

adobe_df = df[adobe]
adobe_df.set_index('job_category', inplace=True)
adobe_df.plot(kind='pie', y='count', legend=True)
```

### 피어슨 상관 계수

-1 ~ 1 사이의 값을 가진다.  
- `0` : x와 y의 연관이 아예 없다.  
- `> 0` : 0보다 크면 1에 가까워질수록 연관성이 더 강해진다. x가 커질수록 y가 커진다는 의미이다.  
- `1` : 확실한 연관성이 있다.
- `< 0` : 0보다 작으면 -1에 가까워질수록 연관성이 더 강해진다. 반대의 경우이기 때문에 x가 커질수록 y가 작아진다는 의미이다.
- `-1` : x와 y는 확실한 반대의 관계이다.

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/exam.csv')

df.corr()  # 숫자 데이터 사이의 상관 계수를 보여줌
```

상관 계수도 DataFrame 형태로 출력된다. 하지만 숫자가 많다보니 한눈에 잘 들어오지 않는다. 이럴 때 **히트맵**을 사용한다.  
히트맵은 상관 계수를 시각화하는 대표적인 방법이다. 상관 계수는 시각화해서 보는 경우가 많다. Seaborn을 이용하면 히트맵을 그릴 수 있다.  

상관 계수의 결과를 Seaborn의 heatmap 메소드에 넘겨주면 된다.  

```python
%matplotlib inline
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/exam.csv')
sns.heatmap(df.corr())
```

색이 밝을수록 상관 계수가 더 높다는 의미이다. 읽기 점수와 쓰기 점수 사이의 상관 관계가 가장 강하다는 것을 한 눈에 알 수 있다.  

`annot=True` 옵션을 추가해주면, 색상 뿐 아니라 숫자도 함께 보여준다.

```python
%matplotlib inline
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/exam.csv')
sns.heatmap(df.corr(), annot=True)
```

### 실습 문제 - 직업 탐구하기 Ⅱ

groupby 문법을 사용해서 '여성 비율'이 높은 순으로 직업을 나열해 보세요.  
DataFrame이 아닌 Series로, 'gender'에 대한 값만 출력되어야 합니다.

*해답 찾아본 문제ㅠ*

```python
import pandas as pd

df = pd.read_csv('data/occupations.csv')

group = df.groupby('occupation')

df.loc[df['gender'] == 'F', 'gender'] = 1
df.loc[df['gender'] == 'M', 'gender'] = 0
df['gender'] = df['gender'].astype(int)  # gender 컬럼의 데이터 타입 변경

group.mean()['gender'].sort_values(ascending=False)
```

### 데이터 합치기
1. inner join
2. left outer join
3. right outer join
4. full outer join

```python
import pandas as pd
```

```python
price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product')  # Product 기준으로 같은 것만 합치기(inner) 
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left') # quantity_df에서 price_df에 있는 내용만 price_df에 합치기(left outer) 
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right') 
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer') 
```
